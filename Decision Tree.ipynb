```python
from sklearn import tree

# Training data
X = [[150, 0], [170, 0], [140, 1], [130, 1]]
y = ["Apple", "Apple", "Orange", "Orange"]

# Create and train the model
model = tree.DecisionTreeClassifier()
model.fit(X, y)

# Predict a new fruit
prediction = model.predict([[145, 1]])

print("Prediction:", prediction[0])
```
